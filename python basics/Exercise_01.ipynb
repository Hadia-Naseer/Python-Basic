Write a Python program to get the largest number from a list

```python
lists = [6, 8, 1, 9, 4]
largest = lists[0]
for number in lists:
    if number > largest:
        largest = number
print("The largest number in the list is:", largest)
```

Write a Python program to remove duplicates from a list

```python
lists = [1, 2, 5, 2, 4, 1, 5, 3]
newlist = []
for item in lists:
    if item not in newlist:
        newlist.append(item)
print("List after removing duplicates:", newlist)
```

Write a Python program to find the index of an item of a tuple.

```python
tuples = (10, 20, 30, 40, 50)
find = 30
if find in tuples:
    index = tuples.index(find)
    print(f"The index of {find} in the tuple is: {index}")
else:
    print(f"{find} is not in the tuple.")
```

Write a Python script to concatenate following dictionaries to create a new one.
Sample Dictionary :
dic1={1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}
Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

```python

dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}
resultdict = dic1.copy()
resultdict.update(dic2)
resultdict.update(dic3)
print("Concatenated Dictionary:", resultdict)
```

Write a Python program to remove a key from a dictionary.

```python
dict1 = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
key = 'b'
if key in dict1:
    removed = dict1.pop(key)
    print(f"Removed key '{key}' with value {removed}.")
else:
    print(f"Key '{key}' not found in the dictionary.")
print("Updated dictionary:", dict1)
```

Write a Python function that takes a list of words and return the longest word and the
length of the longest one.
41
Sample Output:
Longest word: Exercises
Length of the longest word: 9

```python
lists= ["Python", "Exercises", "Code", "Programming", "Functionality"]
longest_word = ""
longest_length = 0
for word in lists:
    if len(word) > longest_length:
        longest_word = word
        longest_length = len(word)
print("Longest word:", longest_word)
print("Length of the longest word:", longest_length)
```

Write a Python program to remove the nth index character from a nonempty string

```python
string = "Hello, World!"
n = 7
if 0 <= n < len(string):
    modified = string[:n] + string[n+1:]
    print("Modified string:", modified)
else:
    print("Error: Index is out of range.")
```
