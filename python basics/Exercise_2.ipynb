Write a Python program which accepts the radius of a circle from the user and compute
the area.
Sample Output :
r = 1.1
Area = 3.8013271108436504

```python

radius = float(input("Enter the radius of the circle: "))
area = 3.141592653589793 * (radius ** 2)
print("Area =", area)
```

Write a Python program to accept a filename from the user and print the extension of
that.

Output : java

```python

filename = input("Enter the filename (with extension): ")
extension = filename.split('.')[-1]
print(extension)
```

Write a Python program to display the examination schedule. (Extract the date from
exam_st_date).
exam_st_date = (11, 12, 2014)
Sample Output : The examination will start from : 11 / 12 / 2014

```python

exam_st_date = (11, 12, 2014)
day, month, year = exam_st_date
print(f"The examination will start from : {day} / {month} / {year}")
```

. Write a Python program to calculate the sum of three given numbers, if the values are
equal then return three times of their sum.

```python

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))
if num1 == num2 == num3:
    result = 3 * (num1 + num2 + num3)
else:
    result = num1 + num2 + num3
print("The result is:", result)
```
