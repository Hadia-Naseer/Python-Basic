## String Slicing
- Extract a subset of characters from a particular strings.
- Accessing specific part of the string by specifying a range of indices.

```
string[start:stop:step]
```

````python
name="Ahmed"
print(name[:])      # output : Ahmed
print(name[0:])      # output : Ahmed
print(name[:5])      # output : Ahmed
print(name[0:5])      # output : Ahmed 
print(name[:len(name)])      # output : Ahmed
print(name[:4])      # output : Ahme
print(name[2:])      # output : med
print(name[0:5:1])      # output : Ahmed ( by default step = 1 )
print(name[0:5:2])      # output : Amd
print(name[::])        # output : Ahmed
print(name[::-1])        # output : demhA  (reverse a string)
print(name[-1])        # output : Ahme
print(name[-0::])        # output : Ahmed
````

## String Methods

````python
name = "My name is .... "

print(name.capitalize()) # First character is capital

print(name.upper()) # Convert all characters into Upper case 

print(name.lower()) # Convert all characters into Lower case

print(name.title()) # Capitalize the first letter of each word

print(name.find("is")) # It returns the index of the first occurrence of the specified value

print(name.endswith("Zahran") or name.endswith("zahran")) # It returns True if the string ends with the specified value

print(name.count("a")) # number of times a specified value occurs in a string

print(name.replace("zahran", "Bilawal"))  # replace special words with new words

print(name.split(" ")) # split the string into a list of words

print(name.strip()) # remove leading and trailing whitespaces

print(name.lstrip()) # remove leading whitespaces in left side 

print(name.rstrip()) # remove trailing whitespaces in right side

print(name.startswith("My") or name.startswith("my")) # It returns True if the string
````

## String format

````python
# using f strings

name = "Ahmad"

age = 22

print(f"My name is {name} and I am {age} years old.")

# using format() method

print("My name is {} and I am {} years old.".format(name, age))

# using % operator

print("My name is %s and I am %d years old." % (name, age))

# using string concatenation

print("My name is " + name + " and I am " + str(age) + " years old.")
````
