Write a Program to Implement Breadth First Search using Python

```python

class Graph:
    def __init__(self):

        self.graph = {}
    def addEdge(self, u, v):

        if u in self.graph:
            self.graph[u].append(v)
        else:
            self.graph[u] = [v]
    def BFS(self, s):

        visited = []

        queue = []

        visited.append(s)
        queue.append(s)
        while queue:

            s = queue.pop(0)
            print(s, end=" ")

            if s in self.graph:
                for i in self.graph[s]:
                    if i not in visited:
                        visited.append(i)
                        queue.append(i)

g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)
print("Following is Breadth First Traversal (starting from vertex 2):")
g.BFS(2)
```
