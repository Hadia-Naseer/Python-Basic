```python
import zipfile
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python

with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    zip_ref.extractall('.')
df = pd.read_csv("winequality-red.csv")
X = df.values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=3)   # Changed from 2 to 3 components for 3D plotting
X_pca = pca.fit_transform(X_scaled)
print("Transformed PCA shape:", X_pca.shape)
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
df.head()
```

```python

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:, 0], X_pca[:, 1],color='blueviolet')
plt.xlabel("Principal Component 1 (PC1)")
plt.ylabel("Principal Component 2 (PC2)")
plt.title("PCA - 2D Visualization")
plt.grid(True)
plt.show()

```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c='rosybrown', s=20)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D PCA of Wine Quality Dataset")

plt.show()
```
