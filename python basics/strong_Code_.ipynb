```python

special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>?/`~"


passw = input('Enter password: ')


if len(passw) == 8:
    has_upper = any(char.isupper() for char in passw)
    has_digit = any(char.isdigit() for char in passw)
    has_special = any(char in special_characters for char in passw)


    if has_upper and has_digit and has_special:
        print("Strong password")
    else:
        if not has_upper:
            print("Password must contain at least one uppercase letter.")
        if not has_digit:
            print("Password must contain at least one digit.")
        if not has_special:
            print("Password must contain at least one special character.")
else:
    print("ALERT .............WEAK PASSWORD: Password must be exactly 8 characters long.")
```
