write a program to propmt the user for hours and rate per hour to compute gross pay

```python
hours = float(input("Enter hours: "))
rate = float(input("Enter rate per hour: "))
gross_pay = hours * rate
print("Gross pay:", gross_pay)
```

write a program which prompts the user for a celsius temperature convert the temperature to fahrenheit and print out converted temperature

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit:", fahrenheit)
```
